```python
# Encode the Target Variable

"""
The target variable (letter) contains alphabetic labels (Aâ€“Z).
We'll encode these into numeric form using LabelEncoder.

"""

from sklearn.preprocessing import LabelEncoder

# Copy the dataset to avoid modifying the original
data = df.copy()

# Encode the target variable 'letter'
encoder = LabelEncoder()
data['letter'] = encoder.fit_transform(data['letter'])

# Display encoded class mapping
label_mapping = dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))
print("Label Encoding Mapping (Alphabet â†’ Number):")
print(label_mapping)
```

```python
data
```

```python
# Feature Normalization
# Since the ANN works best when input features are on a similar scale, weâ€™ll normalize all numeric features using
# Min-Max Scaling (values between 0 and 1).

from sklearn.preprocessing import MinMaxScaler

# Separate features and target
X = data.drop('letter', axis=1)
y = data['letter']

# Normalize features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame for clarity
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print("\nFeature scaling completed.")
display(X_scaled.head())
```

```python
# 2. Model Implementation

from sklearn.model_selection import train_test_split

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# Display shape of splits
print(f"X_train: {X_train.shape}")
print(f"X_test:  {X_test.shape}")
print(f"y_train: {y_train.shape}")
print(f"y_test:  {y_test.shape}")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical

# Since we're performing multi-class classification (26 classes for Aâ€“Z),
# we'll convert the target variable into one-hot encoded format using to_categorical().
# Convert target labels to one-hot encoded format
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)

print(f"Number of classes: {y_train_categorical.shape[1]}")
```

```python
"""
Build the ANN Model:

-> We'll start with a basic ANN architecture:

   Input Layer: 16 neurons (one for each feature)
   Hidden Layer: 1 layer with 32 neurons, ReLU activation
   Output Layer: 26 neurons (softmax activation for multi-class classification)

"""

# Build a basic ANN model
model = Sequential([
    Dense(32, input_dim=X_train.shape[1], activation='relu'),
    Dense(26, activation='softmax')  # 26 output neurons for Aâ€“Z
])

# Compile the model
model.compile(optimizer='adam', 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])

# Model summary
model.summary()
```

```python
# Train the ANN model
# We'll train the model for 20 epochs using the training data and evaluate on validation (test) data.
history = model.fit(
    X_train, y_train_categorical,
    validation_data=(X_test, y_test_categorical),
    epochs=20,
    batch_size=32,
    verbose=1
)
```

```python
# Visualize Training Performance

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

# Accuracy Plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Loss Plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
# Evaluate Model Performance

# Evaluate on test data
test_loss, test_accuracy = model.evaluate(X_test, y_test_categorical, verbose=0)
print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Test Loss: {test_loss:.4f}")
```

```python
"""

Quick Summary Interpretation:

-> Model architecture is simple (just one hidden layer, 32 neurons).
-> Training progress shows steady improvement â€” clear learning curve.
-> Validation accuracy close to training â†’ model generalizes well.
-> Accuracy ~72%, loss = 1.0 â†’ promising baseline, but not yet optimal.
-> No overfitting; network stops improving naturally around epoch 20.

"""
```

```python
# 3. Hyperparameter Tuning

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import RandomizedSearchCV

# Function to create model
def create_model(neurons=32, activation='relu', learning_rate=0.001, dropout_rate=0.0):
    model = Sequential([
        Dense(neurons, input_dim=X_train.shape[1], activation=activation),
        Dropout(dropout_rate),
        Dense(neurons // 2, activation=activation),
        Dense(26, activation='softmax')  # 26 output classes for alphabets (Aâ€“Z)
    ])
    
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Wrap in KerasClassifier for Compatibility with Scikit-learn
model = KerasClassifier(model=create_model, verbose=0)

# Define Hyperparameter Search Space
param_dist = {
    'model__neurons': [32, 64, 128],
    'model__activation': ['relu', 'tanh'],
    'model__learning_rate': [0.001, 0.01],
    'model__dropout_rate': [0.0, 0.2, 0.4],
    'batch_size': [32, 64],
    'epochs': [20, 30]
}

# Perform Random Search for Efficiency
random_search = RandomizedSearchCV(
    estimator=model,
    param_distributions=param_dist,
    n_iter=10,
    cv=3,
    verbose=2
)

# Fit search
random_search_result = random_search.fit(X_train, y_train_categorical)

# Display Best Parameters and Accuracy
print("Best Parameters:", random_search_result.best_params_)
print(f"Best Cross-Validation Accuracy: {random_search_result.best_score_:.4f}")

```

```python
"""

Best Parameters: {'model__neurons': 128, 'model__learning_rate': 0.01, 'model__dropout_rate': 0.2, 
'model__activation': 'relu', 'epochs': 20, 'batch_size': 64}
Best Cross-Validation Accuracy: 0.8796

"""
```

```python
# Evaluate Tuned Model on Test Data
best_params = random_search_result.best_params_

# Build tuned model using best parameters
tuned_model = create_model(
    neurons=best_params['model__neurons'],
    activation=best_params['model__activation'],
    learning_rate=best_params['model__learning_rate'],
    dropout_rate=best_params['model__dropout_rate']
)

# Train tuned model
history_tuned = tuned_model.fit(
    X_train, y_train_categorical,
    validation_data=(X_test, y_test_categorical),
    epochs=best_params['epochs'],
    batch_size=best_params['batch_size'],
    verbose=1
)

# Evaluate tuned model
test_loss_tuned, test_accuracy_tuned = tuned_model.evaluate(X_test, y_test_categorical, verbose=0)
print(f"Tuned Model Test Accuracy: {test_accuracy_tuned:.4f}")
print(f"Tuned Model Test Loss: {test_loss_tuned:.4f}")
```

```python
"""

Performance Interpretation:

-> Test Accuracy (0.8910) â€” The model now correctly classifies nearly 9 out of 10 alphabet samples.
-> Low Test Loss (0.3159) â€” Predictions are much more certain and stable.
-> Smooth Training Curve Expected â€” After tuning, the validation and training curves would stay close, showing
   robust generalization.

Key Insights:

-> Hyperparameter tuning works: Neural network depth, neuron count, and learning rate significantly impacted model 
   quality.
-> Balanced generalization: No overfitting â€” test accuracy and CV accuracy align well (~88â€“89%).
-> Dropout made it stable: The dropout layer helped the model avoid memorizing training data.
-> ANN achieved strong results: This dataset seems well-suited for dense neural architectures.

"""
```

```python
# 4. Evaluation

# Evaluate Both Models (Accuracy, Precision, Recall, F1-Score)
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Predict for both models
y_pred_base = np.argmax(model.predict(X_test), axis=1)
y_pred_tuned = np.argmax(tuned_model.predict(X_test), axis=1)

y_true = np.argmax(y_test_categorical, axis=1)

# Calculate metrics for baseline model
accuracy_base = accuracy_score(y_true, y_pred_base)
precision_base = precision_score(y_true, y_pred_base, average='weighted')
recall_base = recall_score(y_true, y_pred_base, average='weighted')
f1_base = f1_score(y_true, y_pred_base, average='weighted')

# Calculate metrics for tuned model
accuracy_tuned = accuracy_score(y_true, y_pred_tuned)
precision_tuned = precision_score(y_true, y_pred_tuned, average='weighted')
recall_tuned = recall_score(y_true, y_pred_tuned, average='weighted')
f1_tuned = f1_score(y_true, y_pred_tuned, average='weighted')

# Display comparison table
comparison_df = pd.DataFrame({
    "Metric": ["Accuracy", "Precision", "Recall", "F1-score"],
    "Baseline Model": [accuracy_base, precision_base, recall_base, f1_base],
    "Tuned Model": [accuracy_tuned, precision_tuned, recall_tuned, f1_tuned]
})

print("\nModel Performance Comparison:\n")
print(comparison_df.to_string(index=False))
```
