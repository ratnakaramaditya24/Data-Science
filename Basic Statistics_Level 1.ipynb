```python
# Descriptive Analytics 

import pandas as pd
df = pd.read_csv("D:\\PYTHON CLASS\\assignments\\sales_data_with_discounts.csv")
numerical_cols = ['Volume', 'Avg Price', 'Total Sales Value', 
                  'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']
stats = {
    "Column": [],
    "Mean": [],
    "Median": [],
    "Std Dev": [],
    "Mode": []
}
for col in numerical_cols:
    stats["Column"].append(col)
    stats["Mean"].append(df[col].mean())
    stats["Median"].append(df[col].median())
    stats["Std Dev"].append(df[col].std())
    stats["Mode"].append(df[col].mode().values[0]) 

stats_df = pd.DataFrame(stats)
print(stats_df)


# Interpretation:

# Volume is right-skewed since the mean (5.07) is higher than the median (4.00).

# Avg Price, Total Sales Value, and Net Sales Value show high standard deviations.

# Discount Rate (%) has a relatively consistent distribution with a mode of around 5%.

# Outliers may exist in price and sales related columns due to the large gap between mean and median.
```

```python
# Histograms

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("D:\\PYTHON CLASS\\assignments\\sales_data_with_discounts.csv")
numerical_cols = ['Volume', 'Avg Price', 'Total Sales Value', 
                  'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']

sns.set(style="whitegrid")

for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], bins=30, kde=True, color='blue')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
    

# Interpretation:

# Volume: Right-skewed, with most data clustered at lower values.

# Avg Price: Highly skewed to the right.

# Total Sales Value and Net Sales Value: Also show right skewness, many entries are at the lower end with some high-value outliers.

# Discount Rate (%): Fairly bell-shaped and slightly skewed left.

# Discount Amount: Strongly right-skewed with a large number of small values.
```

```python
# Boxplots

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("D:\\PYTHON CLASS\\assignments\\sales_data_with_discounts.csv")

numerical_cols = ['Volume', 'Avg Price', 'Total Sales Value', 
                  'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']

sns.set(style="whitegrid")

for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col], color='lightcoral')
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()


# Interpretation:

# Outliers: Dots or points outside the whiskers.

# Spread: Width of the box shows the interquartile range (IQR).

# Skewness: If the median is not centered in the box, the data is skewed.
```

```python
# Bar Chart Analysis for Categorical Column

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("D:\\PYTHON CLASS\\assignments\\sales_data_with_discounts.csv")

categorical_cols = ['Day', 'SKU', 'City', 'BU', 'Brand', 'Model']

sns.set(style="whitegrid")

for col in categorical_cols:
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, x=col, palette="Set2", order=df[col].value_counts().index)
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


# Interpretation:

# Day: We may find some days have more sales than others.
```

```python
# Standardization of Numerical Variables

import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("D:\\PYTHON CLASS\\assignments\\sales_data_with_discounts.csv")

numerical_cols = ['Volume', 'Avg Price', 'Total Sales Value', 
                  'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']

scaler = StandardScaler()

standardized_data = scaler.fit_transform(df[numerical_cols])

standardized_df = pd.DataFrame(standardized_data, columns=[col + ' (Standardized)' for col in numerical_cols])

comparison_df = pd.concat([df[numerical_cols], standardized_df], axis=1)

comparison_df.head()


# Interpretation:

# Standardization transforms each feature to have mean = 0 and standard deviation = 1.

# This makes variables comparable.
```

```python
# Conversion of Categorical Data into Dummy Variables

import pandas as pd
from sklearn.preprocessing import OneHotEncoder
df = pd.read_csv("D:\\PYTHON CLASS\\assignments\\sales_data_with_discounts.csv")

categorical_cols = ['Day', 'SKU', 'City', 'BU', 'Brand', 'Model']

# Initialize OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, drop=None) 

encoded_array = encoder.fit_transform(df[categorical_cols])

encoded_col_names = encoder.get_feature_names_out(categorical_cols)

encoded_df = pd.DataFrame(encoded_array, columns=encoded_col_names)

encoded_df.head()

# Interpretation

# Why One-Hot Encoding?
# ML algorithms cant work directly with categorical text data. One-hot encoding converts each category into a binary column, 
# allowing algorithms to process the data numerically.
```
